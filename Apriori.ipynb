```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Groceries.csv')
```

```python
basket = []

for i in range(len(data)):
    basket.append(data['items'][i].strip('}').strip('{').split(','))
```

```python
basket#将数据保存在list里
```

```python
#文本编码转换
import itertools
goods = set(itertools.chain(*basket))#不同的商品

num_to_item = {}

item_to_num = {}

for index,item in enumerate(goods):#做映射
    num_to_item[index] = item
    item_to_num[item] = index

```

```python

for i in range(len(basket)):
    for j in range(len(basket[i])):
        basket[i][j] = item_to_num[basket[i][j]]
#在basket做映射
```

```python
#Candidate set1

def build1(basket):
    item1 = set(itertools.chain(*basket))
    return [frozenset([i]) for i in item1]



C1 = build1(basket)
```

```python
def ck_to_lk(dataset , ck , min_support):
    support = {}
    for row in dataset:
        for items in ck:
            if items.issubset(row):
                support[items] = support.get(items,0)+1
    total = len(dataset)
    
    return {k : v / total for k,v in support.items() if v/total >= min_support}

LK1 = ck_to_lk(basket,C1,0.005)
LK1
```

```python
#由频繁1项集组合形成频繁k+1项集

def lk_to_ckset(lk_list):
    ckset = set()
    lk_size = len(lk_list)
    if lk_size > 1:#如果lk只有一个的话，说明无法再向上取候选集了
        k = len(lk_list[0])
        for i,j in itertools.combinations(range(lk_size),2):
            t = lk_list[i]|lk_list[j]
            if len(t)==(k+1):
                ckset.add(t)
    
    return ckset

C2 = lk_to_ckset(list(LK1.keys()))
C2
    
```

```python
#C2里面的东西并不都是所需要的，需要将C2转变为L2才是所需要的
LK2 = ck_to_lk(basket,C2,0.005)
LK2
```

```python
#生成所有的频繁项集

def get_L_all(dataset,min_support):
    c1 = build1(basket)
    L1 = ck_to_lk(basket,c1,min_support)
    
    L_all = L1
    
    Lk = L1
    i=1
    while len(Lk)>1:
        lk_list = list(Lk.keys())
        
        ck = lk_to_ckset(lk_list)
        
        Lk = ck_to_lk(basket,ck,min_support)
        if len(Lk)>0:
            L_all.update(Lk)
        else:
            break
        print(i)
        i+=1
    return L_all

L_all = get_L_all(basket,0.005)
L_all
```

```python
def rule_from_item(item):
    left = []
    for i in range(1,len(item)):
        left.extend(itertools.combinations(item,i))
    return [(frozenset(le),frozenset(item.difference(le))) for le in left]

rule_from_item(frozenset({1,2,3}))
```

```python
#生成关联规则
def rule_from_Lall(L_all,min_confidence):
    rules = []
    for Lk in L_all:
        if len(Lk)>1:
            rules.extend(rule_from_item(Lk))
    
    result = []
    for left,right in rules:
        support = L_all[left|right]
        confidence = support/L_all[left]
        if confidence>min_confidence:
            result.append({'left':left,'right':right,'left|right':confidence})
    
    return result

result = rule_from_Lall(L_all,0.5)
            
result
```

```python
#存储

def back(item):
    li = list(item)
    for i in range(len(li)):
        li[i] = num_to_item[li[i]]
    return li

df = pd.DataFrame(result)
df = df.reindex(['left','right','left|right'],axis =1)
df['left'] = df['left'].apply(back)
df['right'] = df['right'].apply(back)
df
```

```python
df.to_csv('result.csv')
```
